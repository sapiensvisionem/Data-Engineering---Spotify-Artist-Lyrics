```python
# Libaries
import pip  
import urllib2
from bs4 import BeautifulSoup
import csv
```

```python
# Parsing function
import urllib

def make_url(base_url , *res):
    url = base_url
    for r in res:
        url = '{}/{}'.format(url, r)
    return url+'.html'
```

```python
# List of songs and singers
with open('C:/Users/jihun/Desktop/songlist.csv','rb') as csvfile:
    songlist = list(csv.reader(csvfile))
```

```python
# Create url list based on songs and singers
urllist=[]
for pair in songlist:
    urllist.append(make_url('http://azlyrics.com/lyrics',pair[1],pair[0]))
```

```python
# scrape lyrics
data=[] 
import time
for pg in urllist[416:]:
    try: 
        page = urllib2.urlopen(pg)
        soup = BeautifulSoup(page,'html.parser')
        lyrics = soup.find_all('div',attrs={'class': None})[1].get_text()
        data.append(lyrics)
        time.sleep(5)
    except:
        data.append("URL Does Not Exist#$%^")
data
```

```python
# Save into CSV File with open('lyrics.csv','a') as csv_file:
    writer=csv.writer(csv_file)
    for lyrics in data:
        writer.writerow(lyrics)
```

```python
data = [lyric.replace("\n","") for lyric in data]
```

```python
data = [lyric.replace("\\","") for lyric in data]
```
